```python
# Loader data
import pandas as pd
from datetime import datetime
import seaborn as sns
import matplotlib.pyplot as plt
# Henter datasÃ¦t
data = pd.read_csv("../1_opstart/Data/data_2003_to_May_2018.csv")
data["Dato"] = pd.to_datetime(data['Date'], format="%m/%d/%Y") # Make the data datetime
data['year']= data['Dato'].dt.year # Make a variable for year
data["week"]= data["Dato"].dt.day_name()
data["month"]= data["Dato"].dt.month_name()
data["Hour"] = pd.to_datetime(data["Time"], format="%H:%M").dt.hour
data["Minute"] = pd.to_datetime(data["Time"], format="%H:%M").dt.minute
data["12_hours"] = [datetime.strftime(datetime.strptime(str(t), "%H"), "%I %p") for t in data["Hour"] ]
data["Weekday_Hour"] = data["week"] + "-" + data["12_hours"] 
data = data[data["year"]!=2018]

weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
hour_order = [datetime.strftime(datetime.strptime(str(t), "%H"), "%I %p") for t in range(24)]
weekday_hour_order = [weekday + "-" + hour for weekday in weekday_order for hour in hour_order]
```

```python
# Choosing the data. I select the two months of Oktober, November and December in 2010
data_1 = data[data["year"]==2010]
data_1 = data_1[(data_1["month"]=="November") | (data_1["month"]=="December") | (data_1["month"]=="October")]
# I then choose the hour of 13
data_1 = data_1[(data_1["Hour"]==13)]

# Now i make the plot 
data_1['Time'] = pd.to_datetime(data_1['Time'])
data_1["minute"] = data_1["Time"].dt.minute

# Set the color palette to black and white
sns.set_palette("Greys", desat=1)
# Create the stripplot
sns.stripplot(data=data_1, x="minute", jitter= 0.2, alpha=0.5, color="black")

# Set the x-axis to show only the minutes
plt.xticks(range(0, 59, 5))

#Add labels and titles
plt.title('Crimes at 13:00-13:59:59 in October-Dec 2010')
plt.xlabel('Time')
plt.ylabel('Number of Crimes')

plt.show()
```

In the plot we see a tendency for crimes happening at "round" numbers. We see alot of crime happening at exactly 13:0, 13:15, and 13:30.

```python
import numpy as np
# I look into Assault and Fraud in 2010 and within october, november and december
data_2 = data[data["year"]==2010]
#data_2 = data_2[(data_2["month"]=="November") | (data_2["month"]=="December") | (data_2["month"]=="October")]
data_ass = data_2[data_2["Category"]=="ROBBERY"]
data_fra = data_2[data_2["Category"]=="PROSTITUTION"]

# extract the latitude data into a numpy array
latitudes_ass = np.array(data_ass['Y'])
latitude_fra = np.array(data_fra["Y"])

# Constructs subplot 
fig, axs = plt.subplots(1, 2, sharey=True, tight_layout=True)
# use numpy.histogram to generate the histogram data
axs[0].hist(latitudes_ass, bins=50, alpha=0.5, color='b')
axs[0].set_xlabel('Latitude')
axs[0].set_ylabel('Frequency')
axs[0].set_title('Histogram of Robbery latitude')

# plot the longitude histogram on the second subplot
axs[1].hist(latitude_fra, bins=50, alpha=0.5, color='r')
axs[1].set_xlabel('Latitude')
axs[1].set_ylabel('Frequency')
axs[1].set_title('Histogram of Prostitution latitude')

plt.show()
```
